```
#@title Crop images into 896 * 896 * 3 for 3024 * 4032 * 3 in dataset one
def data_processing(get_path, post_path, grade):
  images = glob.glob(get_path+ '/*.JPG')
  k=0
  for i in images:  
    print(i)
    img = mpimg.imread(i) 
    if img.shape[0]==3024 and img.shape[1]==4032:
      crop_image = img[750:1646, 1000:1896]
      crop_image=crop_image[0:896, 0:896]    
    k=k+1
    print(str(grade)+'('+str(k)+')')
    print(crop_image.shape)
    pil_img = Image.fromarray(crop_image)
    pil_img.save(post_path+ '/'+ str(grade)+'('+str(k)+')'+'.JPG')
  return k
print("set 1 grade 1: "+str(data_processing('/content/drive/My Drive/FYR/pre-process/1', '/content/drive/My Drive/FYR/pre-processed/1', 1)))
print("set 1 grade 2: "+str(data_processing('/content/drive/My Drive/FYR/pre-process/2', '/content/drive/My Drive/FYR/pre-processed/2', 2)))
print("set 1 grade 3: "+str(data_processing('/content/drive/My Drive/FYR/pre-process/3', '/content/drive/My Drive/FYR/pre-processed/3', 3)))
```

```
#@title Create overall dataset for train
def create_blocks(get_path, post_path, grade):
  images = glob.glob(get_path+ '/*.JPG')
  m=0
  for i in images:
    img = mpimg.imread(i) 
    for k in range(0, 7):
      for l in range(0, 7):
        img_crop=img[k*128:(k+1)*128, l*128:(l+1)*128]
        m=m+1
        pil_img = Image.fromarray(img_crop)
        pil_img.save(post_path+ '/'+ str(grade)+'('+str(m)+')'+'.JPG')
  return m
print("grade 1 count: "+str(create_blocks('/content/drive/My Drive/FYR/all/1', '/content/drive/My Drive/FYR/blocked_all/1', 1)))
print("grade 2 count: "+str(create_blocks('/content/drive/My Drive/FYR/all/2', '/content/drive/My Drive/FYR/blocked_all/2', 2)))
print("grade 3 count: "+str(create_blocks('/content/drive/My Drive/FYR/all/3', '/content/drive/My Drive/FYR/blocked_all/3', 3)))    
```

```
#@title Overall data pre-processing for train (RGB)
def load_train_data(path):
  images = glob.glob(path + '/*.JPG')
  random.shuffle(images)
  x = []
  y= []
  t=0
  r=0
  for i in images:
    temp = cv2.imread(i, cv2.IMREAD_COLOR)
    temp=np.array(temp)
    temp = temp.astype('float32')
    temp =temp/ 255
    x.append(temp)
    if int(i[39:40])==1:
      r=0
    elif int(i[39:40])==2:
      r=1
    else:
      r=2
    y.append(r)
    t=t+1
    print("path: "+str(i))
    print("count: "+str(t)+" "+"category: "+str(r)+" "+"size: "+str(temp.shape)) 
  x=np.array(x)
  y=np.array(y)
  x = x.reshape(len(x), 128, 128, 3)
  y = np_utils.to_categorical(y, 3)
  np.save('/content/drive/My Drive/FYR/train_data/x_train.npy',x)
  np.save('/content/drive/My Drive/FYR/train_data/y_train.npy',y)
  print('size of x: ',str(x.shape))
  print('size of y: ',str(y.shape))
load_train_data('/content/drive/My Drive/FYR/merged-all') 
```

```
#@title Overall data pre-processing for test (RGB)
def load_test_data(path, grade):
  images = glob.glob(path + '/*.JPG')
  random.shuffle(images)
  x = []
  t=0
  for i in images:
    temp = cv2.imread(i, cv2.IMREAD_COLOR)
    temp=np.array(temp)
    temp = temp.astype('float32')
    temp =temp/ 255
    x.append(temp)
    t=t+1
    print("path: "+str(i))
    print("count: "+str(t)+" "+"size: "+str(temp.shape)) 
  x=np.array(x)
  x = x.reshape(len(x), 128, 128, 3)
  np.save('/content/drive/My Drive/FYR/test_data/x_test_cat_'+str(grade)+'.npy',x)
  return str(x.shape)
print('size of x for grade 1: '+load_test_data('/content/drive/My Drive/FYR/merged-test-all/1', 1))
print("###########################################################################")
print('size of x for grade 2: '+load_test_data('/content/drive/My Drive/FYR/merged-test-all/2', 2)) 
print("###########################################################################")
print('size of x for grade 3: '+load_test_data('/content/drive/My Drive/FYR/merged-test-all/3', 3)) 
```
